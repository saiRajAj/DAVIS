Write a function in python to read the content from a text file "ABC.txt" line by line and display the same on screen.

```python
file = open("ABC.txt", "r")
lines = file.readlines()
for line in lines:
  print(line)
file.close()
```

Write a function in Python to count and display the total number of words in a text file “ABC.txt”

```python
file = open('ABC.txt', 'r')
read_file = file.read()
words = read_file.split()
print(len(words))
file.close()
```

Write a function in Python to count uppercase character in a text file “ABC.txt”

```python
file = open("ABC.txt", "r")
text = file.read()
count = 0
for char in text:
  if char.isupper():
    count += 1
print(count)
file.close()
```

Write a function display_words() in python to read lines from a text file "story.txt", and display those words, which are less than 4 characters.

```python
file = open('story.txt', 'w')
file.write('Hello this is the story.txt in the directory...!')
file = open("story.txt", "r")
lines = file.readlines()
for line in lines:
  words = line.split()
  for word in words:
    if len(word) < 4:
      print(word)

file.close()
```

Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.

```python
try :
  print(10/0)
except ZeroDivisionError:
  print("Zero Division Error")
```

Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical

```python
try:
  num1_str = input("Enter the first number: ")
  try:
    num1 = int(num1_str)
  except ValueError:
    raise TypeError("First input is not numerical")

  num2_str = input("Enter the second number: ")
  try:
    num2 = int(num2_str)
  except ValueError:
    raise TypeError("Second input is not numerical")

  print(num1 / num2)
except TypeError as e:
  print(f"Error: {e}")
except ZeroDivisionError:
  print("Error: Cannot divide by zero.")
```

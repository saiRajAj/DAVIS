Using max() and min() functions display the maximum and minimum of 5 random
numbers.

```python
import random
random_numbers = [random.randint(1, 100) for _ in range(5)]
print(f"The 5 random numbers are: {random_numbers}")
max_number = max(random_numbers)
print(f"The maximum number is: {max_number}")
min_number = min(random_numbers)
print(f"The minimum number is: {min_number}")
```

Write a Python program to count the occurrences of each word in a
given sentence
string = “To change the overall look of your document. To change the look
available in the gallery”

```python
import re
string = "To change the overall look of your document. To change the look available in the gallery"
cleaned_string = re.sub(r'[^a-zA-Z\s]', '', string).lower()
words = cleaned_string.split()
word_counts = {}
for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1
print("Word occurrences:")
for word, count in word_counts.items():
    print(f"{word}: {count}")
```

Write a Python program to reverse words in a string String = “Deeptech Python Training”

```python
text = "Welcome to python Training"
vowels = "aeiou"
vowel_counts = {}
text_lower = text.lower()
for char in text_lower:
    if char in vowels:
        vowel_counts[char] = vowel_counts.get(char, 0) + 1

print(f"Original text: '{text}'")
print("Vowel counts:")
for vowel, count in vowel_counts.items():
    print(f"{vowel}: {count}")
```

Write a Python program to Count all letters, digits, and special
symbols from the given string
Input = “P@#yn26at^&i5ve”
Output: Chars = 8 Digits = 2 Symbol = 3

```python
input_string = "P@#yn26at^&i5ve"

char_count = 0
digit_count = 0
symbol_count = 0

for char in input_string:
    if char.isalpha():
        char_count += 1
    elif char.isdigit():
        digit_count += 1
    else:
        symbol_count += 1

print(f"Input: '{input_string}'")
print(f"Output: Chars = {char_count} Digits = {digit_count} Symbol = {symbol_count}")
```

Write a Python program to remove duplicate characters of a given
string.
Input = “String and String Function”
Output: String and Function

```python
input_string = "String and String Function"
words = input_string.split()
unique_words_ordered = []
seen_words = set()

for word in words:
    if word not in seen_words:
        unique_words_ordered.append(word)
        seen_words.add(word)

output_string = " ".join(unique_words_ordered)
print(f"Input: '{input_string}'")
print(f"Output: '{output_string}'")
```

Write a Python program to count Uppercase, Lowercase, special
character and numeric values in a given string

Input = “Hell0 W0rld ! 123 * # welcome to pYtHoN”
Output:
UpperCase : 5
LowerCase : 18
NumberCase : 5
SpecialCase : 11

```python
input_string = "Hell0 W0rld ! 123 * # welcome to pYtHoN"

uppercase_count = 0
lowercase_count = 0
number_count = 0
special_count = 0

for char in input_string:
    if char.isupper():
        uppercase_count += 1
    elif char.islower():
        lowercase_count += 1
    elif char.isdigit():
        number_count += 1
    else:
        special_count += 1

print(f"Input: '{input_string}'")
print(f"UpperCase : {uppercase_count}")
print(f"LowerCase : {lowercase_count}")
print(f"NumberCase : {number_count}")
print(f"SpecialCase : {special_count}")
```

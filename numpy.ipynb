# **Numpy**

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
```

```python
print(arr)
```

```python
print(arr[1:3])
```

```python
print(arr[:4])
```

```python
print(arr[2:])
```

```python
print(arr[:-3])
```

**Function in Numpy**

```python
arr = np.arange(1,6)
print(arr)
```

```python
print(np.zeros(5))
```

```python
print(np.ones(5))
```

```python
marks = np.array([70, 80, 50, 90, 60])
print('Sum : ',np.sum(marks))
print('Mean : ',np.mean(marks))
print('Max : ', np.max(marks))
```

#2D Numpy Array

```python
arr = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])
print(arr)
```

```python
print(arr.shape)
```

#2d Array accessing Elements

```python
print(arr[0,1])
```

```python
print(arr[2,1])
```

```python
print(arr[0:2,0:2])
```

```python
print(arr[1:3,1:3])
```

```python
print(arr.ndim)
```

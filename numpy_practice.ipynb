# **1D NumPy Array Questions**

Write a program to create a 1D NumPy array with five elements entered by the user.


```python
import numpy as np
```

```python
arr = np.array([int(input()) for i in range(5)])
```

```python
print(arr)
```

Write a program to find the length of a 1D NumPy array.

```python
len = len(arr)
print(len)
```

Write a program to access the first, middle, and last element of a NumPy array.


```python
print(arr[0], arr[len//2], arr[len-1])
```

Write a program to slice a NumPy array from index 2 to 5.


```python
print(arr[2:6])
```

Write a program to print all elements of a NumPy array using a loop.


```python
for element in arr:
  print(element, end=" ")
```

Write a program to add 10 to each element of a NumPy array.


```python
print(arr + 10)
```

Write a program to find the sum and average of elements in a 1D NumPy array.

```python
print('Sum : ', np.sum(arr))
print('Average : ', np.average(arr))
```

Write a program to find the maximum and minimum value in a NumPy array.


```python
print('Maximum : ', np.max(arr))
print('Minimum : ', np.min(arr))
```

Write a program to reverse a NumPy array using slicing.

```python
print(arr[::-1])
```

Write a program to check whether a specific element exists in a NumPy array.


```python
element = int(input('Enter Element to search : '))
print(element in arr)
```

# **2D NumPy Array Questions**

Write a program to create a 2D NumPy array of size 3Ã—3 and display it.

```python
arr = np.array([[int(input()) for i in range(3)] for j in range(3)])
```

```python
print(arr)
```

Write a program to find the shape of a 2D NumPy array.

```python
print('Shape of array is : ', arr.shape)
```

Write a program to access a specific element from a 2D array using row and column index.

```python
element = int(input('Enter Element to search : '))
print(np.where(arr == element))
```

Write a program to extract the first row and first column from a 2D NumPy array.


```python
print('Row : ', arr[0:1])
print('Column : ', arr[:,0])
```

Write a program to slice a 2D NumPy array to get a sub-matrix.


```python
print(arr[1:3, 1:3])
```

Write a program to calculate the sum of all elements in a 2D NumPy array.


```python
print(np.sum(arr))
```

Write a program to calculate row-wise sum of a 2D NumPy array.


```python
print(np.sum(arr, axis=1))
```

Write a program to calculate column-wise average of a 2D NumPy array.


```python
print(np.average(arr, axis=0))
```

Write a program to multiply each element of a 2D NumPy array by 2.


```python
print(arr * 2)
```

Write a program to find the maximum value from each row of a 2D NumPy array.


```python
print(np.max(arr, axis=1))
```

Write a program to find the largest number in a list.

```python
list1 = []
r = int(input("Enter the size of list: "))
for i in range(r):
  list1.append(int(input(f"Enter numbers in list: ")))
print(max(list1))
```

Write a program to count how many times a particular element appears in a list.

```python
list1 = []
r = int(input("Enter the size of list: "))
for i in range(r):
  list1.append(int(input("Enter numbers in list: ")))
element = int(input('Enter the element you want to count : '))
print(list1.count(element))
```

Write a program to add a new element to a list and remove an existing element.

```python
list1 = []
r = int(input("Enter the size of list: "))
for i in range(r):
  list1.append(int(input("Enter numbers in list: ")))
print(list1)
n = int(input('Enter the Element you want to add in list: '))
o = int(input('Enter the Element you want to remove from list: '))
i = list1.index(o)
list1[i] = n
print(list1)
```

Write a program to store student details using a dictionary and display them.

```python
dict1 = {}
stud = int(input('Enter total no of students : '))
for i in range(1,stud+1):
  name = input(f'Enter name of student {i} : ')
  marks = int(input(f'Enter marks of student {i} : '))
  dict1[name] = marks
print(dict1)
```

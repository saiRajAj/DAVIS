Write a Python program to get the largest and smallest number from a list without
builtin functions.

```python
list1 = [10,20,30,50,60,40]
min = list1[0]
max = list1[0]
for i in list1:
  if i < min:
    min = i
  if i > max:
    max = i
print(min)
print(max)
```

Write a Python program to find duplicate values from a list and display those.

```python
list1 = [10,20,30,10,30,50]
duplicates = []
seen = set()

for item in list1:
  if item in seen:
    duplicates.append(item)
  else:
    seen.add(item)

print("Duplicate values:", list(set(duplicates)))
```

Write a Python program to split a given list into two parts where the length of the first
part of the list is given.
Original list:
[1, 1, 2, 3, 4, 4, 5, 1]
Length of the first part of the list: 3
Splitted the said list into two parts:
([1, 1, 2], [3, 4, 4, 5, 1])

```python
original_list = [1, 1, 2, 3, 4, 4, 5, 1]
length_first_part = 3
```

```python
first_part = original_list[:length_first_part]
second_part = original_list[length_first_part:]

print("Original list:", original_list)
print("Length of the first part of the list:", length_first_part)
print("Splitted the said list into two parts:", (first_part, second_part))
```

Write a Python program to traverse a given list in reverse order, and print the
elements with the original index.
Original list:
['red', 'green', 'white', 'black']
Traverse the said list in reverse order:
black
white
green
red

```python
original_list = ['red', 'green', 'white', 'black']
print("Traverse the said list in reverse order:")
for i in range(len(original_list) - 1, -1, -1):
    print(f"Element: {original_list[i]}, Original Index: {i}")
```

Write a Python program and calculate the mean of the below dictionary.
test_dict = {"A" : 6, "B" : 9, "C" : 5, "D" : 7, "E" : 4}

```python
test_dict = {"A" : 6, "B" : 9, "C" : 5, "D" : 7, "E" : 4}
total_sum = sum(test_dict.values())
count = len(test_dict)
mean_value = total_sum / count

print("Mean of the dictionary values:", mean_value)
```

3. Write a Python program to calculate the sum of the numbers in a given tuple.
Input: tuples_list = [(1, 2), (3, 4), (5, 6)]

```python
tuples_list = [(1, 2), (3, 4), (5, 6)]
total_sum = 0
for tup in tuples_list:
    for num in tup:
        total_sum += num
print("Sum of the numbers in the tuples:", total_sum)
```

1. Write a Python program to Get Only unique items from two sets.

2. Write a Python program to Return a set of elements present in Set A or B, but
not both.

set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
Output:
{20, 70, 10, 60}

3. Write a Python program to Check if two sets have any elements in common. If
yes, display the common elements.
Input:

set1 = {10, 20, 30, 40, 50}
set2 = {60, 70, 80, 90, 10}

Output:
{10}

```python
set1 = {10, 20, 30, 40, 50}
set2 = {60, 70, 80, 90, 10}
common_elements = set1.intersection(set2)

if common_elements:
    print(common_elements)
else:
    print("No common elements found.")
```

Python program to check leap year

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
year = int(input("Enter a year: "))
if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

A toy vendor supplies three types of toys: Battery Based Toys, Key-based
Toys, and Electrical Charging Based Toys. The vendor gives a discount of
10% on orders for battery-based toys if the order is for more than Rs. 1000.
On orders of more than Rs. 100 for key-based toys, a discount of 5% is
given, and a discount of 10% is given on orders for electrical charging based
toys of value more than Rs. 500. Assume that the numeric codes 1,2 and 3
are used for battery based toys, key-based toys, and electrical charging based
toys respectively. Write a program that reads the product code and the order
amount and prints out the net amount that the customer is required to pay
after the discount.

```python
def calculate_net_amount(product_code, order_amount):
    net_amount = order_amount

    if product_code == 1:
        if order_amount > 1000:
            net_amount = order_amount * 0.90
    elif product_code == 2:
        if order_amount > 100:
            net_amount = order_amount * 0.95
    elif product_code == 3:
        if order_amount > 500:
            net_amount = order_amount * 0.90
    else:
        print("Invalid product code. Please enter 1, 2, or 3.")
        return -1

    return net_amount

try:
    product_code = int(input("Enter product code (1 for Battery, 2 for Key-based, 3 for Electrical): "))
    order_amount = float(input("Enter the order amount (Rs.): "))

    final_amount = calculate_net_amount(product_code, order_amount)
    if final_amount != -1:
        print(f"The net amount to pay after discount is: Rs. {final_amount:.2f}")
except ValueError:
    print("Invalid input. Please enter valid numbers for product code and order amount.")
```

A transport company charges the fare according to following table:
Distance Charges

1-50 8 Rs./Km

51-100 10 Rs./Km

> 100 12 Rs/Km

```python
def calculate_fare(distance):
    if distance <= 0:
        return 0
    elif distance <= 50:
        return distance * 8
    elif distance <= 100:
        return (50 * 8) + ((distance - 50) * 10)
    else:
        return (50 * 8) + (50 * 10) + ((distance - 100) * 12)
try:
    distance_travelled = float(input("Enter the distance traveled in Km: "))

    total_fare = calculate_fare(distance_travelled)
    if distance_travelled <= 0:
        print("Distance must be a positive number.")
    else:
        print(f"The total fare for {distance_travelled:.2f} Km is: Rs. {total_fare:.2f}")
except ValueError:
    print("Invalid input. Please enter a valid number for distance.")
```

Write a python program to check whether a number is palindrome or
not?

```python
num = int(input("Enter an integer: "))

original_num = num
reversed_num = 0

while num > 0:
    remainder = num % 10
    reversed_num = (reversed_num * 10) + remainder
    num = num // 10

if original_num == reversed_num:
    print(f"{original_num} is a palindrome.")
else:
    print(f"{original_num} is not a palindrome.")
```

Write a python program finding the factorial of a given number using
a while loop.

```python
num = int(input("Enter a non-negative integer: "))

factorial = 1
count = num

if num < 0:
    print("Factorial does not exist for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    while count > 0:
        factorial *= count
        count -= 1
    print(f"The factorial of {num} is {factorial}.")
```

Accept numbers using input() function until the user enters 0. If user
input 0 then break the while loop and display the sum of all the
numbers.

```python
total_sum = 0

print("Enter numbers (enter 0 to stop and see the sum):")

while True:
    try:
        num = int(input("Enter a number: "))
        if num == 0:
            break
        total_sum += num
    except ValueError:
        print("Invalid input. Please enter an integer.")

print(f"The sum of all entered numbers is: {total_sum}")
```

Python program to check if the given string is a palindrome

```python
input_string = input("Enter a string: ")

processed_string = input_string.replace(" ", "").lower()

reversed_string = processed_string[::-1]

if processed_string == reversed_string:
    print("String is a palindrome.")
else:
    print("String is not a palindrome.")
```

Python program to check if a given number is an Armstrong number

```python
num = int(input("Enter a number: "))
original_num = num
num_digits = len(str(num))

sum_of_powers = 0

temp_num = num
while temp_num > 0:
    digit = temp_num % 10
    sum_of_powers += digit ** num_digits
    temp_num //= 10

if original_num == sum_of_powers:
    print(f"{original_num} is an Armstrong number.")
else:
    print(f"{original_num} is not an Armstrong number.")
```

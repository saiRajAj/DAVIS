Write a program to read the all the content of the file and count the no of lines present in it

```python
file = open("om.txt",'r')
count = 0
while True:
    line = file.readline()
    count += 1
    if not line:
        break
    print(line)
file.close()
print(count-1)
```

Write a program to handle error when a file does not exist using exception handling

```python
try:
  file = open('sai.txt','r')
  print(file.read())
except FileNotFoundError:
  print("File does not exist")
```

Explain the difference between write ('w') mode and append ('a') mode with a small code example

```python
print("Using Write Mode...!\n")
file = open('data.txt','r')
print('Content of file before operation :')
print(file.read())
print('\n')
file.close()
file = open("data.txt",'w')
file.write('Hello....!')
file.close()
file = open("data.txt",'r')
print('Content in file after operation : ')
print(file.read())
print('\n\n------------------------------------------------------------\n\n')
file.close()
print("Using append mode...!\n")
file = open('data.txt','r')
print('Content of file before operation :')
print(file.read())
print('\n')
file.close()
file = open("data.txt",'a')
file.write('\nHello')
file.close()
file = open("data.txt",'r')
print('Content in file after operation : ')
print(file.read())
file.close()
```

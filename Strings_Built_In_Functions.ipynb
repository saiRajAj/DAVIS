Write a program to take two numbers from the user and print their sum using type conversion.

```python
num1_str = input("Enter the first number: ")
num2_str = input("Enter the second number: ")

num1 = int(num1_str)
num2 = int(num2_str)

sum_of_numbers = num1 + num2

print("The sum of the two numbers is:", sum_of_numbers)
```

Write a program to find the maximum of two numbers using a built-in function.

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

max_of_numbers = max(num1, num2)

print("The maximum of the two numbers is:", max_of_numbers)
```

Write a program to count the number of characters in a string without using len().

```python
str = input("Enter a string: ")
count = 0
for char in str:
    count += 1
print("The number of characters in the input string is:", count)
```

Write a program to reverse a string.

```python
string = input("Enter a string to reverse: ")
reversed_string = string[::-1]
print("The reversed string is:", reversed_string)
```

Write a program to check whether a given string is a palindrome.

```python
string = input("Enter a string to reverse: ")
reversed_string = string[::-1]
if string == reversed_string:
    print("The input string is a palindrome")
else:
    print("The input string is not a palindrome")
```

Write a program to check whether an email ID contains the symbol @.

```python
string = input("Enter Email ID : ")
s = bool(string.find('@'))
if s == True:
    print("The input string contains the symbol @")
else:
    print("The input string does not contain the symbol @")
```

Write a program to count the number of vowels in a string.

```python
string = input("Enter a string: ")
count = 0
for char in string:
    if char in "aeiouAEIOU":
        count += 1
print("The number of vowels in the input string is:", count)
```

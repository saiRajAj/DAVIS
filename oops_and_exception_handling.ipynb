What is encapsulation? How does Python achieve data hiding? Explain with an example.


```python
class details:
  def __init__(self, name, age):
    self.__name = name
    self.__age = age

  def get_name(self):
    return self.__name

  def get_age(self):
    return self.__age

o = details("John", 36)
print(o.get_name())
print(o.get_age())
```

What is inheritance? Write a Python program to show single inheritance.

```python
class base:
  def __init__(self, name, age):
    self.name = name
    self.age = age

class derived(base):
  def __init__(self, name, age):
    base.__init__(self, name, age)

o = derived("John", 36)
print(o.name)
print(o.age)
```

Write a Python program to handle division by zero and invalid input using exception handling.

```python
a = int(input('Enter a Number : '))
b = int(input('Enter a Number : '))
try:
  c = a / b
  print(c)
except ZeroDivisionError as e:
  print(e)
```

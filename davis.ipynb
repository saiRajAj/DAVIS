Write a program to check whether a given character is a vowel or a consonant.

```python
char = input("Enter a character: ")

char_lower = char.lower()

if char_lower in ('a', 'e', 'i', 'o', 'u'):
  print(f"The character '{char}' is a vowel.")
else:
  print(f"The character '{char}' is a consonant.")
```

Write a program to calculate electricity bill based on units consumed.

```python
units = int(input("Enter the number of units consumed: "))
bill = 0

if units <= 50:
  bill = units * 10
elif units <= 150:
  bill = (50 * 10) + ((units - 50) * 15)
else:
  bill = (50 * 10) + (100 * 15) + ((units - 150) * 20)

print(f"Your electricity bill for {units} units is: ${bill}")
```

Write a program to check whether a year is a leap year or not.

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
  print(f"{year} is a leap year.")
else:
  print(f"{year} is not a leap year.")
```

```python
salary = float(input("Enter employee's annual salary: "))
bonus = 0

if salary < 30000:
  bonus_rate = 0.05
elif salary <= 50000:
  bonus_rate = 0.08
else:
  bonus_rate = 0.10

bonus = salary * bonus_rate

print(f"For a salary of ${salary:.2f}, the bonus is: ${bonus:.2f}")
```

Write a program to count the number of digits in a given number.

 Write a program to take marks of 5 students and calculate the average.

```python
total_marks = 0
num_students = 5

for i in range(num_students):
  marks = float(input(f"Enter marks for student {i+1}: "))
  total_marks += marks

average_marks = total_marks / num_students

print(f"The total marks for {num_students} students is: {total_marks}")
print(f"The average marks for {num_students} students is: {average_marks:.2f}")
```

Write a program to take 5 numbers from the user and count positive and negative numbers.

```python
positive_count = 0
negative_count = 0
zero_count = 0

for i in range(5):
  num = int(input(f"Enter number {i+1}: "))
  if num > 0:
    positive_count += 1
  elif num < 0:
    negative_count += 1
  else:
    zero_count += 1

print(f"You entered {positive_count} positive numbers.")
print(f"You entered {negative_count} negative numbers.")
print(f"You entered {zero_count} zeros.")
```
